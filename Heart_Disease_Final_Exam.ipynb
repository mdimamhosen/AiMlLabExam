## Task 5 and 6: Machine Learning Models and Evaluation
Three models were trained: Logistic Regression, Random Forest, and Support Vector Machine.

- Logistic Regression works like a simple scoring system and predicts the chance of disease.
- Random Forest builds many decision trees and combines their votes.
- SVM finds the best boundary between disease and no disease patients.

SVM was used because this is a medical classification task. The table below shows accuracy, precision, recall, F1-score, ROC-AUC, and confusion matrix values.

Metric meaning: Accuracy is total correct predictions. Precision shows how correct the disease predictions are. Recall shows how many real disease patients were found. F1-score balances precision and recall. ROC-AUC shows how well the model separates disease and no disease classes.
